# SQL Notes and Commands

## Creating and Managing Databases and Tables

### Creating a Database
To create a new database:
```sql
CREATE DATABASE database_name;
```
Example:
```sql
CREATE DATABASE customers;
```

### Dropping a Database
To delete an existing database:
```sql
DROP DATABASE database_name;
```
Example:
```sql
DROP DATABASE customers;
```

### Selecting a Database
To select a database to use:
```sql
USE database_name;
```
Example:
```sql
USE customers;
```

### Showing Databases
To list all databases:
```sql
SHOW DATABASES;
```

## Creating and Managing Tables

### Creating a Table
To create a new table within a database:
```sql
CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    ...
);
```
Example:
```sql
CREATE TABLE customer_info (
    id INT,
    first_name VARCHAR(20),
    last_name VARCHAR(20),
    age INT
);
```

### Dropping a Table
To delete an existing table:
```sql
DROP TABLE table_name;
```
Example:
```sql
DROP TABLE customer_info;
```

### Showing Tables
To list all tables in the selected database:
```sql
SHOW TABLES;
```

## Inserting Data into Tables

### Inserting Data
To insert data into a table:
```sql
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```
Examples:
```sql
INSERT INTO customer_info (id, first_name, last_name, age) 
VALUES (1, 'Uzair', 'Hussain', 21);

INSERT INTO customer_info (id, first_name, last_name, age) 
VALUES (2, 'Razi', 'Rizwan', 20);
```

## Querying Data from Tables

### Selecting Data
To retrieve data from a table:
```sql
SELECT * FROM table_name;
```
Example:
```sql
SELECT * FROM customer_info;
```
