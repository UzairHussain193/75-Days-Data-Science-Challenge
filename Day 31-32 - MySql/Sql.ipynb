# SQL Notes and Commands

## Creating and Managing Databases and Tables

### Creating a Database
To create a new database:
```sql
CREATE DATABASE database_name;
```
Example:
```sql
CREATE DATABASE customers;
```

### Dropping a Database
To delete an existing database:
```sql
DROP DATABASE database_name;
```
Example:
```sql
DROP DATABASE customers;
```

### Selecting a Database
To select a database to use:
```sql
USE database_name;
```
Example:
```sql
USE customers;
```

### Showing Databases
To list all databases:
```sql
SHOW DATABASES;
```

## Creating and Managing Tables

### Creating a Table
To create a new table within a database:
```sql
CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    ...
);
```
Example:
```sql
CREATE TABLE customer_info (
    id INT,
    first_name VARCHAR(20),
    last_name VARCHAR(20),
    age INT
);
```

### Dropping a Table
To delete an existing table:
```sql
DROP TABLE table_name;
```
Example:
```sql
DROP TABLE customer_info;
```

### Showing Tables
To list all tables in the selected database:
```sql
SHOW TABLES;
```

## Inserting Data into Tables

### Inserting Data
To insert data into a table:
```sql
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```
Examples:
```sql
INSERT INTO customer_info (id, first_name, last_name, age) 
VALUES (1, 'Uzair', 'Hussain', 21);

INSERT INTO customer_info (id, first_name, last_name, age) 
VALUES (2, 'Razi', 'Rizwan', 20);
```

## Querying Data from Tables

### Selecting Data
To retrieve data from a table:
```sql
SELECT * FROM table_name;
```
Example:
```sql
SELECT * FROM customer_info;
```

# Day 32 - SQL Notes

## SQL Null Values

### Checking for Null Values
To select records where a column has a `NULL` value:
```sql
SELECT * FROM table_name WHERE column_name IS NULL;
```
Example:
```sql
SELECT * FROM customer_info WHERE salary IS NULL;
```

### Checking for Non-Null Values
To select records where a column does not have a `NULL` value:
```sql
SELECT * FROM table_name WHERE column_name IS NOT NULL;
```
Example:
```sql
SELECT * FROM customer_info WHERE salary IS NOT NULL;
```

## SQL Update Statement

### Updating Data
To update existing records in a table:
```sql
UPDATE table_name SET column1 = value1, column2 = value2, ... WHERE condition;
```
Example:
```sql
UPDATE customer_info SET salary = 68000 WHERE id = 4;
```

## SQL Delete Statement

### Deleting Data
To delete existing records from a table:
```sql
DELETE FROM table_name WHERE condition;
```
Example:
```sql
DELETE FROM customer_info WHERE id = 3;
```

## SQL Alter Table

### Adding a Column
To add a new column to an existing table:
```sql
ALTER TABLE table_name ADD column_name datatype;
```
Example:
```sql
ALTER TABLE customer_info ADD email VARCHAR(20);
```

### Modifying/Altering a Column
To modify an existing column in a table:
```sql
ALTER TABLE table_name MODIFY column_name new_datatype;
```
Example:
```sql
ALTER TABLE customer_info MODIFY dobb YEAR;
```

### Dropping a Column
To delete a column from an existing table:
```sql
ALTER TABLE table_name DROP COLUMN column_name;
```
Example:
```sql
ALTER TABLE customer_info DROP COLUMN dobb;
```
