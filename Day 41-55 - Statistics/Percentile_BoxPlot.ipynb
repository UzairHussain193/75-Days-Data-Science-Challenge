```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define datasetset
dataset = [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107,10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
plt.hist(dataset)
```

```python
# Z-score
outliers = []
def detect_outliers(dataset):
    
    threshold = 3 # 3 standard deviations away from the mean
    mean = np.mean(dataset) # mean
    std = np.std(dataset)   # standard deviation
    
    for i in dataset: # iterate over dataset
        z_score = (i-mean)/std # calculate z-score
        if np.abs(z_score) > threshold: # if z-score is greater than threshold
            outliers.append(i) # add to outliers list
    return outliers # return outliers
```

```python
detect_outliers(dataset)
```

## IQR - Inter Quartile Range
1. Sort the data
2. Calculate Q1 (25%)
3. Calculate Q3 (75%)
4. Calculate IQR (Q3-Q1)
5. Find the Lower Fence --> L.F = Q1 - 1.5(IQR)
6. Find the Upper Fence --> U.F = Q3 + 1.5(IQR)

```python
# 1. Sorting the dataset
dataset = sorted(dataset)
```

```python
dataset
```

```python
#2. Calculate 1st and 3rd quartile
q1, q3 = np.percentile(dataset,[25,75])
print(q1,q3)
```

```python
# 3. Find IQR
iqr = q3 - q1
print(iqr)
```

```python
# 4. Find lower Fence value and the higher Fence value
lower_fence_val = q1 - (1.5 * iqr)
upper_fence_val = q3 + (1.5 * iqr)

print("Lower Fence Value: ", lower_fence_val)
print("Upper Fence Value: ", upper_fence_val)
```

```python
# 5. Identify outliers
outliers = []
for i in dataset:
    if i < lower_fence_val or i > upper_fence_val:
        outliers.append(i)

print("Outliers: ", outliers)
```
